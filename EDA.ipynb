# Outlier detection
 1. Box plot/Whisker plot
 2. Scatter plot
 3. Z-scores / Standard deviation method (3*SD): Data points beyond +/-3*SD are treated as outliers.
 4. IQR (interquartile range): 

```python
sns.boxplot(df['bmi'])
```

```python
Q1 = df['bmi'].quantile(0.25)
Q3 = df['bmi'].quantile(0.75)

IQR = Q3 - Q1
print(IQR)

```

```python
UW = Q3+(1.5*IQR)
LW = Q1-(1.5*IQR)
print(UW, LW)
```

```python
df.loc[df['bmi']<(LW)] | df.loc[df['bmi']>(UW)]
```
