```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

import seaborn as sns
```

```python
iris = load_iris()
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
df.head()
```

```python
y = iris.target
```

```python
df.corr()
```

```python
# The above dataset has heavy multicollinearity and hence applying PCA.
# Steps to apply PCA
# 1. Apply standard scalar
# 2. Apply PCA
```

```python
df = StandardScaler().fit_transform(df)
```

```python
df
```

```python
pca = PCA(n_components=3)
```

```python
p_comp = pca.fit_transform(df)
```

```python
pca_df = pd.DataFrame(p_comp, columns=['PCA1', 'PCA2', 'PCA3'])
```

```python
pca_df.corr()
```

```python
pca.explained_variance_ratio_
```

```python
x_train, x_test, y_train, y_test = train_test_split(pca_df, y, test_size=0.2, random_state=50)
```

```python
lr = LogisticRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
lr_pred = lr.predict(x_test)
```

```python
accuracy_score(y_test, lr_pred)
```
